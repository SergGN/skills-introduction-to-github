Создание двумерного массива размерностью 100 на 5 (или 100 массивов по 5 элементов в каждом) со случайными числами от 0 до 9 включительно
 и запись его в переменную random_array:

```python
import numpy as np
import pandas as pd
random_array = np.random.randint(low=0,high=20,size=(100,6))
display(random_array)
```

Из получившегося numpy-массива random_array сделаем датафрейм df:


```python
df = pd.DataFrame(random_array,columns=['a','b','c','d','e','f'])
display(df.head())
```
